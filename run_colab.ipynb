```python
import os
import zipfile

source_dir = 'sinkhorn-vision-kd'
dest_dir = './'
zip_file_name = 'sinkhorn-vision-kd.zip'

# Check if the source directory exists
if not os.path.exists(source_dir) or not os.path.isdir(source_dir):
    print(f"Directory '{source_dir}' not found. Attempting to extract from {zip_file_name}...")
    if os.path.exists(zip_file_name):
        try:
            with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
                zip_ref.extractall('./')
            print(f"✅ Successfully extracted {zip_file_name}.")
        except zipfile.BadZipFile:
            print(f"❌ Error: {zip_file_name} is not a valid zip file. Please check the uploaded file.")
        except Exception as e:
            print(f"❌ An unexpected error occurred during extraction: {e}")
    else:
        print(f"❌ Error: {zip_file_name} not found. Please upload the project zip file.")

# Now, attempt to move contents if the source directory exists (either it was there, or just extracted)
if os.path.exists(source_dir) and os.path.isdir(source_dir):
    # Move all contents from source_dir to dest_dir
    print(f"Moving all contents from '{source_dir}' to '{dest_dir}'.")
    for item in os.listdir(source_dir):
        # Use shutil.move for robustness, handles files and directories correctly across filesystems
        # If shutil is not available, os.rename is used, but it's less robust for directories across devices
        try:
            os.rename(os.path.join(source_dir, item), os.path.join(dest_dir, item))
        except OSError as e:
            print(f"Warning: Could not move {item} using os.rename. Error: {e}. Trying shutil.move...")
            import shutil
            shutil.move(os.path.join(source_dir, item), os.path.join(dest_dir, item))

    # Optionally, remove the now empty source directory
    os.rmdir(source_dir)
    print(f"Moved all contents from '{source_dir}' to '{dest_dir}'.")
else:
    print(f"Directory '{source_dir}' still not found after extraction attempt. Skipping move operation.")

print("New project structure:")
print(os.listdir('.'))
```

```python
import os
import zipfile
from google.colab import files

project_zip = 'project_source_only.zip'
exclude_dirs = {'data', '.venv', '.git', '__pycache__', 'sample_data'}

print(f"📦 Creating {project_zip} (excluding large data/venv folders)...")

with zipfile.ZipFile(project_zip, 'w', zipfile.ZIP_DEFLATED, allowZip64=True) as zf:
    for root, dirs, files_list in os.walk('.'):
        # Modify dirs in-place to skip excluded directories
        dirs[:] = [d for d in dirs if d not in exclude_dirs]

        for file in files_list:
            if file == project_zip or file == 'full_project_backup.zip':
                continue
            file_path = os.path.join(root, file)
            # Create an archive name relative to the current directory
            arcname = os.path.relpath(file_path, '.')
            zf.write(file_path, arcname)

print(f"✅ Created {project_zip}")
files.download(project_zip)
```
